```python
import pandas as pd

def detect_outliers_iqr(df, columns):
    """
    Detects outliers in specified columns of a DataFrame using the IQR method and prints an outlier summary.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    columns (list): List of column names to check for outliers.

    Returns:
    dict: Dictionary containing outlier indices and counts for each column.
    """
    outlier_summary = {}

    print("Outlier Detection Summary (Using IQR Method):")
    print("-" * 50)
    
    for column in columns:
        if column not in df.columns:
            print(f"Column '{column}' not found in the DataFrame.")
            continue

        # Calculate Q1, Q3, and IQR
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_indices = outliers.index.tolist()
        outlier_count = len(outlier_indices)

        # Store results in the summary
        outlier_summary[column] = {
            "Lower Bound": lower_bound,
            "Upper Bound": upper_bound,
            "Outlier Count": outlier_count,
            "Outlier Indices": outlier_indices,
        }

        print(f"Column: {column}")
        print(f"  Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
        print(f"  Outlier Count: {outlier_count}")
        print("-" * 50)

    return outlier_summary

# Specify numeric columns
columns = [
    "BALANCE", "BALANCE_FREQUENCY", "PURCHASES", "ONEOFF_PURCHASES",
    "INSTALLMENTS_PURCHASES", "CASH_ADVANCE", "PURCHASES_FREQUENCY",
    "ONEOFF_PURCHASES_FREQUENCY", "PURCHASES_INSTALLMENTS_FREQUENCY",
    "CASH_ADVANCE_FREQUENCY", "CASH_ADVANCE_TRX", "PURCHASES_TRX",
    "CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS", "PRC_FULL_PAYMENT", "TENURE"
]

# Example usage
outlier_summary = detect_outliers_iqr(df, columns)
```

```python
import pandas as pd

def treat_outliers(df, columns, method='capping'):
    """
    Treats outliers in specified columns of a DataFrame by either capping or removing them.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    columns (list): List of column names to treat for outliers.
    method (str): The method to treat outliers - 'capping' or 'removal'.

    Returns:
    pd.DataFrame: A new DataFrame with outliers treated.
    """
    df_copy = df.copy()  # Create a copy to preserve the original data
    print(f"Treating outliers using the {method} method...")

    for column in columns:
        if column not in df_copy.columns:
            print(f"Column '{column}' not found in the DataFrame.")
            continue

        # Calculate Q1, Q3, and IQR
        Q1 = df_copy[column].quantile(0.25)
        Q3 = df_copy[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if method == 'capping':
            # Cap the outliers to the boundary values
            df_copy[column] = df_copy[column].apply(
                lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x)
            )
            print(f"Column: {column} - Capped values below {lower_bound:.2f} and above {upper_bound:.2f}")

        elif method == 'removal':
            # Remove the rows with outliers
            df_copy = df_copy[(df_copy[column] >= lower_bound) & (df_copy[column] <= upper_bound)]
            print(f"Column: {column} - Removed outlier rows outside the range ({lower_bound:.2f}, {upper_bound:.2f})")

        else:
            print(f"Invalid method '{method}' specified. Please choose either 'capping' or 'removal'.")
            return df

    return df_copy

# Example usage:
columns = [
    "BALANCE", "BALANCE_FREQUENCY", "PURCHASES", "ONEOFF_PURCHASES",
    "INSTALLMENTS_PURCHASES", "CASH_ADVANCE", "PURCHASES_FREQUENCY",
    "ONEOFF_PURCHASES_FREQUENCY", "PURCHASES_INSTALLMENTS_FREQUENCY",
    "CASH_ADVANCE_FREQUENCY", "CASH_ADVANCE_TRX", "PURCHASES_TRX",
    "CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS", "PRC_FULL_PAYMENT", "TENURE"
]

# Assuming df is your original DataFrame
df_capped = treat_outliers(df, columns, method='capping')
df_removed = treat_outliers(df, columns, method='removal')
```

```python
def detect_skewness_all_columns(df):
    """
    Detects skewness in all numeric columns of a DataFrame.

    Parameters:
    df (pd.DataFrame): The input DataFrame.

    Returns:
    pd.DataFrame: A DataFrame summarizing skewness for all numeric columns.
    """
    # Select numeric columns only
    numeric_columns = df_capped.select_dtypes(include=['number']).columns
    skewness_summary = []

    print("Skewness Detection Summary for All Numeric Columns:")
    print("-" * 60)

    for column in numeric_columns:
        # Calculate skewness
        skewness_value = df_capped[column].skew()

        # Categorize skewness
        if abs(skewness_value) < 0.5:
            skewness_category = "Symmetrical"
        elif 0.5 <= abs(skewness_value) <= 1:
            skewness_category = "Moderately Skewed"
        else:
            skewness_category = "Highly Skewed"

        # Print and store results
        print(f"Column: {column}")
        print(f"  Skewness: {skewness_value:.2f} ({skewness_category})")
        print("-" * 60)

        skewness_summary.append({
            "Column": column,
            "Skewness Value": skewness_value,
            "Skewness Category": skewness_category
        })

    # Convert results to a DataFrame for further analysis
    skewness_df = pd.DataFrame(skewness_summary)
    return skewness_df

# Example usage:
# Assuming `df` is your DataFrame
skewness_results = detect_skewness_all_columns(df)
print(skewness_results)
```

```python
import numpy as np
import pandas as pd

def treat_skewness(df_capped, columns):
    """
    Treats skewness in all numeric columns of a DataFrame using transformations.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    columns (list): List of numeric columns to treat for skewness.

    Returns:
    pd.DataFrame: A new DataFrame with skewness treated.
    """
    df_capped_copy = df_capped.copy()  # Create a copy to preserve the original data
    print("Treating skewness in all numeric columns...")

    for column in columns:
        if column not in df_capped_copy.columns:
            print(f"Column '{column}' not found in the DataFrame.")
            continue

        # Calculate skewness
        skewness_value = df_capped_copy[column].skew()

        # Apply transformations based on skewness
        if skewness_value > 1:  # Highly positively skewed
            df_capped_copy[column] = np.log1p(df_capped_copy[column].clip(lower=0))  # Log transformation for positive values
            print(f"Column: {column} - Log Transformation applied (Highly Positively Skewed).")
        elif skewness_value < -1:  # Highly negatively skewed
            df_capped_copy[column] = np.log1p(df_capped_copy[column].max() + 1 - df_capped_copy[column])  # Reverse Log Transformation
            print(f"Column: {column} - Reverse Log Transformation applied (Highly Negatively Skewed).")
        elif 0.5 < abs(skewness_value) <= 1:  # Moderately skewed
            df_capped_copy[column] = np.cbrt(df_capped_copy[column])  # Cube root transformation
            print(f"Column: {column} - Cube Root Transformation applied (Moderately Skewed).")

    return df_capped_copy

# Example usage:
columns = [
    "BALANCE", "BALANCE_FREQUENCY", "PURCHASES", "ONEOFF_PURCHASES",
    "INSTALLMENTS_PURCHASES", "CASH_ADVANCE", "PURCHASES_FREQUENCY",
    "ONEOFF_PURCHASES_FREQUENCY", "PURCHASES_INSTALLMENTS_FREQUENCY",
    "CASH_ADVANCE_FREQUENCY", "CASH_ADVANCE_TRX", "PURCHASES_TRX",
    "CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS", "PRC_FULL_PAYMENT", "TENURE"
]

# Assuming df is your original DataFrame
df_treated = treat_skewness(df_capped, columns)

# You can compare the skewness of the original and treated DataFrame
print("\nOriginal DataFrame Skewness:")
print(df_capped[columns].skew())

print("\nTreated DataFrame Skewness:")
print(df_treated[columns].skew())
```

```python
import numpy as np

def select_features_based_on_correlation(df, upper_threshold=0.9, lower_threshold=-0.5):
    """
    Select features based on correlation matrix by removing highly or poorly correlated features.
    
    Parameters:
    df (pd.DataFrame): DataFrame for which features are selected.
    upper_threshold (float): Correlation threshold above which features will be considered highly correlated.
    lower_threshold (float): Correlation threshold below which features will be considered poorly correlated.
    
    Returns:
    selected_columns (list): List of selected columns after removing highly or poorly correlated features.
    """
    # Compute the correlation matrix
    corr_matrix = df.corr()
    
    # Get the upper triangle of the correlation matrix (excluding diagonal)
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Initialize a list to track columns to drop
    to_drop = []
    
    # Find columns with correlations above the upper threshold or below the lower threshold
    for column in upper_triangle.columns:
        if any(upper_triangle[column] > upper_threshold) or any(upper_triangle[column] < lower_threshold):
            to_drop.append(column)
    
    # Select columns that are not in the to_drop list
    selected_columns = [column for column in df.columns if column not in to_drop]
    
    return selected_columns

# Apply the function to select features for df_treated
selected_columns = select_features_based_on_correlation(df_treated, upper_threshold=0.9, lower_threshold=-0.5)

# Show the selected features
print("Selected Features:", selected_columns)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

def plot_tsne(df_scaled, perplexity=30, random_state=42):
    """
    Plots t-SNE visualization for scaled features with enhanced visual appeal.

    Parameters:
    df_scaled (pd.DataFrame): Scaled DataFrame.
    perplexity (int): Perplexity parameter for t-SNE.
    random_state (int): Random seed for reproducibility.
    """
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=random_state)
    tsne_result = tsne.fit_transform(df_scaled)

    # Generate random colors for clusters
    colors = sns.color_palette("husl", n_colors=len(df_scaled))
    plt.figure(figsize=(12, 8))
    
    # Plot with vibrant colors
    scatter = plt.scatter(
        tsne_result[:, 0], tsne_result[:, 1], 
        c=range(len(df_scaled)), cmap='Spectral', alpha=0.8, edgecolor="k", s=50
    )
    plt.colorbar(scatter, label="Sample Index")
    plt.title("t-SNE Visualization (Enhanced)", fontsize=16, fontweight="bold")
    plt.xlabel("t-SNE Dimension 1", fontsize=12)
    plt.ylabel("t-SNE Dimension 2", fontsize=12)
    plt.grid(color="gray", linestyle="--", alpha=0.6)
    plt.show()

# Example usage
plot_tsne(df_scaled)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

def plot_tsne_with_colors(df_scaled, labels=None, perplexity=30, random_state=42):
    """
    Plots t-SNE visualization for scaled features with enhanced coloring by clusters or labels.

    Parameters:
    df_scaled (pd.DataFrame): Scaled DataFrame.
    labels (array-like, optional): Cluster or category labels for coloring.
    perplexity (int): Perplexity parameter for t-SNE.
    random_state (int): Random seed for reproducibility.
    """
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=random_state)
    tsne_result = tsne.fit_transform(df_scaled)

    # Create a color palette
    if labels is not None:
        unique_labels = sorted(set(labels))
        palette = sns.color_palette("husl", n_colors=len(unique_labels))
        color_map = {label: palette[i] for i, label in enumerate(unique_labels)}
        colors = [color_map[label] for label in labels]
    else:
        # Default to a sequential color map
        colors = sns.color_palette("husl", n_colors=len(df_scaled))

    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(
        tsne_result[:, 0], tsne_result[:, 1],
        c=colors, alpha=0.8, edgecolor="k", s=50
    )
    
    if labels is not None:
        # Create a legend for clusters
        for label, color in color_map.items():
            plt.scatter([], [], color=color, label=f"Cluster {label}", edgecolor="k", s=50)
        plt.legend(title="Clusters", fontsize=10, loc="best")
    
    plt.title("t-SNE Visualization with Clusters", fontsize=16, fontweight="bold")
    plt.xlabel("t-SNE Dimension 1", fontsize=12)
    plt.ylabel("t-SNE Dimension 2", fontsize=12)
    plt.grid(color="gray", linestyle="--", alpha=0.6)
    plt.show()

# Example usage:
# Assuming df_scaled is a DataFrame of scaled features and cluster_labels is an array of cluster IDs
# plot_tsne_with_colors(df_scaled, labels=cluster_labels)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def plot_elbow_method(df_scaled, max_clusters=10, random_state=42):
    """
    Plots the Elbow Method curve to find the optimal number of clusters.

    Parameters:
    df_scaled (pd.DataFrame): Scaled DataFrame for clustering.
    max_clusters (int, optional): Maximum number of clusters to evaluate. Defaults to 10.
    random_state (int, optional): Random seed for reproducibility. Defaults to 42.
    """
    inertia = []
    range_n_clusters = range(1, max_clusters + 1)

    # Compute inertia for each number of clusters
    for n_clusters in range_n_clusters:
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
        kmeans.fit(df_scaled)
        inertia.append(kmeans.inertia_)

    # Find the "elbow" point (optional, approximate)
    optimal_clusters = 4  # Placeholder, manually determined or use a heuristic method

    # Plotting the elbow curve
    plt.figure(figsize=(10, 6))
    plt.plot(range_n_clusters, inertia, marker='o', linestyle='--', color='b')
    plt.title('Elbow Method for Optimal Clusters', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Clusters', fontsize=12)
    plt.ylabel('Within-Cluster Sum of Squares (WCSS)', fontsize=12)
    plt.xticks(range_n_clusters)
    plt.grid(color='gray', linestyle='--', alpha=0.7)

    # Highlight the optimal number of clusters
    if optimal_clusters in range_n_clusters:
        plt.axvline(x=optimal_clusters, color='red', linestyle='--', label=f'Optimal Clusters = {optimal_clusters}')
        plt.scatter(optimal_clusters, inertia[optimal_clusters - 1], color='red', s=100, label='Elbow Point')

    plt.legend()
    plt.show()

# Example usage
plot_elbow_method(df_scaled, max_clusters=10)
```
